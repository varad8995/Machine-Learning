```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head(140)
```

```python
df.shape
```

## univariate analysis


```python
df_setosa = df.loc[df["Species"]=="Iris-setosa"]
df_virginica = df.loc[df["Species"]=="Iris-virginica"]
df_versicolour = df.loc[df["Species"]=="Iris-versicolour"]
```

```python
plt.plot(df_setosa["PetalLengthCm"],np.zeros_like(df_setosa["PetalLengthCm"]),"o")
plt.plot(df_virginica["PetalLengthCm"],np.zeros_like(df_virginica["PetalLengthCm"]),"o")
plt.plot(df_versicolour["PetalLengthCm"],np.zeros_like(df_versicolour["PetalLengthCm"]),"o")
plt.xlabel("Petal Length")
plt.show
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue = "Species",size=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue = "Species",size=5).map(plt.scatter,"SepalLengthCm","PetalLengthCm").add_legend();
plt.show()
```

## multivariate analysis

```python
sns.pairplot(df,hue = "Species",size=5)
```
