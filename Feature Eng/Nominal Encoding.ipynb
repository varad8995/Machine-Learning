## One Hot Encoding - variables with many categories

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("mercedesbenz.csv" ,usecols=["X1","X2","X3","X4","X5"])
```

```python
data.head()
```

```python
##lets have a look at how many labels each varible has
for col in data.columns:
    print(col ,':',len(data[col].unique()) , "labels")
```

```python
##let's examine how many columns we will obtain after one hot encoding these varibles
pd.get_dummies(data,drop_first = True).shape
```

we can see that from just 6 intial categorical variables,  we end up with 106 new varibles.


```python
##let's find the top 10 most frequet categories for varible x2

data.X2.value_counts().sort_values(ascending = False).head(20)
```

```python
##let's make a list with the most frequent frequent categories of the varible
top_10 = [X for X in data.X2.value_counts().sort_values(ascending = False).head(20).index]
```

```python
top_10
```

```python
for label in top_10:
    data[label]= np.where(data['X2'] == label , 1 , 0)
    
data[['X2']+top_10].head(40)    
    
```

```python
##get whole set of dummy varibles, for all the categorical variables

def one_hot_topx(df,variable,top_x_labels):
    ##fuction to create the dumy varibles for the most frequent labels
    ## we can vary the number of most frequent label that we encode
    
    for label in top_x_labels:
        df[variable + "-"+label]=np.where(data[variable]==label,1,0)
        
##read the data again
data = pd.read_csv("mercedesbenz.csv",usecols = ["X1","X2","X3","X4","X5","X6"])

##encode x2 into 10 most frequqent categoris
one_hot_topx(data,"X2",top_10)
data.head()
```

```python
top_10 = [X for X in data.X1.value_counts().sort_values(ascending = False).head(20).index]
one_hot_topx(data,"X1",top_10)
data.head()
```

```python
top_10 = [X for X in data.X3.value_counts().sort_values(ascending = False).head(20).index]
one_hot_topx(data,"X3",top_10)
data.head()
```

```python
top_10 = [X for X in data.X4.value_counts().sort_values(ascending = False).head(20).index]
one_hot_topx(data,"X4",top_10)
data.head()
```

```python
top_10 = [X for X in data.X5.value_counts().sort_values(ascending = False).head(20).index]
one_hot_topx(data,"X5",top_10)
data.head()
```
